```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
df
```

```python
df['duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
df['duration'] = df['duration'].dt.total_seconds()/60
```

```python
df
```

```python
df = df[['passenger_count','payment_type', 'fare_amount', 'trip_distance', 'duration']]
```

```python
df.isnull().sum()
```

```python
(4896/len(df))*100
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
df['passenger_count'] = df['passenger_count'].astype('int64')
df['payment_type'] = df['payment_type'].astype('int64')
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace = True) 
```

```python
df['passenger_count'].value_counts(normalize = True)          #parameter (normalize = True) gives the percentage count
```

```python
df['payment_type'].value_counts(normalize = True)
```

```python
df = df[df['payment_type']<3]
df = df[(df['passenger_count']>0)&(df['passenger_count']<6)]
```

```python
df.shape
```

```python
df['payment_type'].replace([1,2], ['Card','Cash'], inplace = True)
```

```python
df
```

```python
df.describe()
```

```python
df = df[df['fare_amount']>0]
df = df[df['trip_distance']>0]
df = df[df['duration']>0]
```

```python
plt.boxplot(df['fare_amount'])
```

```python
for col in['fare_amount','trip_distance','duration']:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3-q1
    
    lower_bound = q1 -1.5*IQR
    upper_bound = q3 +1.5*IQR
    
    df  =df[(df[col]>=lower_bound) & (df[col]<=upper_bound)]
df
```

```python
plt.figure(figsize = (12,5))

plt.subplot(1,2,1)
plt.title('Distribution of fare amount')
plt.hist(df[df['payment_type'] == 'Card']['fare_amount'], histtype = 'barstacked', bins = 20, edgecolor = 'k', color = '#FA643F', label = 'Card' )
plt.hist(df[df['payment_type'] == 'Cash']['fare_amount'], histtype = 'barstacked', bins = 20, edgecolor = 'k', color = '#FFBCAB', label = 'Cash' )
plt.legend()


plt.subplot(1,2,2)
plt.title('Distribution of trip distance')
plt.hist(df[df['payment_type'] == 'Card']['trip_distance'], histtype = 'barstacked', bins = 20, edgecolor = 'k', color = '#FA643F', label = 'Card' )
plt.hist(df[df['payment_type'] == 'Cash']['trip_distance'], histtype = 'barstacked', bins = 20, edgecolor = 'k', color = '#FFBCAB', label = 'Cash' )
plt.legend()
plt.show()
```

```python
df.groupby('payment_type').agg({'fare_amount':['mean','std'], 'trip_distance':['mean','std']})
```

```python
plt.title('Preference for payment type')
plt.pie(df['payment_type'].value_counts(normalize = True),labels = df['payment_type'].value_counts().index,
    startangle = 90, shadow = True, autopct = '%1.1f%%', colors = ['#FA643F','#FFBCAB'])
plt.show()
```

```python
df.groupby(['payment_type', 'passenger_count'])[['passenger_count']].count()
```

```python
passenger_count = df.groupby(['payment_type', 'passenger_count'])[['passenger_count']].count()
passenger_count.rename(columns = {'passenger_count': 'count'}, inplace = True)
passenger_count.reset_index(inplace = True)
```

```python
passenger_count['percantage'] = (passenger_count['count']/passenger_count['count'].sum())*100
passenger_count
```
